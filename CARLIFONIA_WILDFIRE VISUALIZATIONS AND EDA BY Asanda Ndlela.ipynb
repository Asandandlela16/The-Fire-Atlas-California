```python
#NOW THAT I HAVE CHANGED THE REQUIRED DATATYTPES , WE WILL RENAME OUR COLUMNS TO BE MORE READABLE BY STANDARDIZING THEM.

carlifonia_dataset.columns=["incident id","date","location","areas burned","homes destroyed","businesses destroyed","vehicles damaged","injuries",
                            "fatalities","estimated financial loss(million $)","causes"]
```

```python
carlifonia_dataset.info()

```

```python
#NOW LET US LOOK AT HOW WE COULD RE INDEX OUR DATASET , WE DO THIS IN ORDER TO HAVE NEAT PERSENTATION AND DATES ARE ALWAYS
# BEING INDEXED.

carlifonia_dataset.reset_index(drop=True, inplace=True)

```

```python
#SAY WE WANT TO FIND THE AVERAGE ESTIMATED FINANCIAL LOSS PER LOCATION(MILLION DOLLARS)ANDF AVERAGE VEHICLES DAMAGED PER LOCATION.

carlifonia_dataset.groupby("location")["estimated financial loss(million $)"].mean().round(0)
```

```python
carlifonia_dataset.groupby("location")["vehicles damaged"].mean().round(0)
```

```python
#SO NOW WE CAN SEE which location WAS THE MOST and least AFFECTED IN TERMS OF FINANCIAL LOSS AND  CARS BEING DESTROYED BY THE FIRE.
```

```python
# NOW  LET US FULLY MOVE TO THE VISUALIZATIONS

carlifonia_dataset[carlifonia_dataset["location"]=="Butte County"].plot
```

```python
carlifonia_dataset.groupby("location")["vehicles damaged"].mean().round(0).plot(kind="bar" , title="Average vehicles damaged per location" )
plt.ylabel("Vehicles damaged")
plt.show()
```

```python
carlifonia_dataset.groupby("location")["estimated financial loss(million $)"].mean().round(0).plot(kind="bar"  , title=" Estimated financial location per region")
plt.xlabel("location")
plt.ylabel("Finacial loss( Million $)")
plt.show()
```

```python
carlifonia_dataset["estimated financial loss(million $)"].plot(kind="hist" , bins=10 , title="Loss")
plt.xlabel("Income")
plt.show()
```

```python
#Let us look at the loss distribution with Seaborn , we use kde for a smooth curve .

sns.histplot(carlifonia_dataset["estimated financial loss(million $)"] , kde=True )
plt.title("Loss distribution")
plt.show()
```

```python
#LET US TRY TO SHOW THE SPREAD OF INJURIES NOW

sns.boxplot(x=carlifonia_dataset["injuries"])
plt.title("Injuries spread amount")
plt.show()
```
